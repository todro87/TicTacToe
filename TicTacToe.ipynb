```python
from IPython.display import clear_output

board_state=[" "," "," "," "," "," "," "," "," "]
board_state_hidden=[-1,-1,-1,-1,-1,-1,-1,-1,-1]
players={"x": 1, "o": 2}

def init_board():
    
    global board_state
    global board_state_hidden
    
    board_state=[" "," "," "," "," "," "," "," "," "]
    board_state_hidden=[-1,-1,-1,-1,-1,-1,-1,-1,-1]
    
def get_position():
    
    pos=input("Choose position:")
    
    x=0
    
    while x==0:
        if board_state[int(pos)-1]!=" ":
            print("Position is not empty, try another one.")
            x=0
            pos=input()
        else:
            x=1
            return int(pos)

def set_position(pos, marker):
    global board_state
    global board_state_hidden
    
    if pos>=1 and pos<=9:
            board_state[pos-1]=marker
            board_state_hidden[pos-1]=0
    else:
        print("Position should be between 1 and 9.")

        
def win_check(players, marker, board_state):
    
    # rows
    if board_state[0:3]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    elif board_state[3:6]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    elif board_state[6:9]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    # diagonals
    elif [board_state[0], board_state[4], board_state[8]]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    elif [board_state[2], board_state[4], board_state[6]]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    # columns
    elif [board_state[0], board_state[3], board_state[6]]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    elif [board_state[1], board_state[4], board_state[7]]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    elif [board_state[2], board_state[5], board_state[8]]==[marker,marker,marker]:
        print("Player {} wins!".format(players[marker]))
        return 0
    else:
        return 1
        
def print_board():
    clear_output()
    print("   |   |")
    print(" "+str(board_state[6])+" | "+str(board_state[7])+" | "+str(board_state[8])+" ")
    print("   |   |")
    print("-----------")
    print("   |   |")
    print(" "+str(board_state[3])+" | "+str(board_state[4])+" | "+str(board_state[5])+" ")
    print("   |   |")
    print("-----------")
    print("   |   |")
    print(" "+str(board_state[0])+" | "+str(board_state[1])+" | "+str(board_state[2])+" ")
    print("   |   |")
    
def switch_player(player_num, marker):
    global active_player
    
    active_player[0]=player_num
    active_player[1]=marker
    
    return active_player
    
def start_game():
    
    global players
    
    init_board()
    print_board()
    i=0
    check=1
    
    while sum(board_state_hidden)!=0:
        if (i+1)%2!=0:
            print("Player 1: x")
            marker="x"
        else:
            print("Player 2: o")
            marker="o"

        pos=get_position()
        set_position(pos, marker)
        check=win_check(players, marker, board_state)
        print_board()
        
        if check==0:
            print("End of game!")
            if marker=="x":
                print("Player 1 wins!")
            else:
                print("Player 2 wins!")
            break
        
        i+=1
        
        
```
